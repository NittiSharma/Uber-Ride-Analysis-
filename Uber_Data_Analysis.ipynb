```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/UberDataset - UberDataset.csv")
```

```python
dataset.info()
```

```python
# Data Processing - PURPOSE Column has so many NULL values
dataset ['PURPOSE'].fillna("NOT", inplace = True)
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')
```

```python
dataset['START_DATE']
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')
dataset ['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce')
```

```python
# Splitting the Start_Date to date and time column and then converting the time into
#four categories i.e Morning , Afternoon, Evening and Night

from datetime import datetime
dataset ['date'] = pd.DatetimeIndex(dataset ['START_DATE']).date
dataset['time'] = pd.DatetimeIndex(dataset['START_DATE']).hour
# changing the categories of Day and Night

dataset['day-night'] = pd.cut( x = dataset['time'], bins = [0,10,15,19,24],
                              labels = ['Morning', 'Afternoon', 'Evening','Night'])
```

```python
# Once done with creating the new columns , we can now drop the rows with null values
dataset.dropna(inplace = True )
```

```python
# It  is important to remove the duplicates
dataset.drop_duplicates(inplace = True)
```

```python
dataset
```

```python
# DATA VISUALIZATION
# Checking the unique values in dataset of the columns with object datatype

obj = (dataset.dtypes == 'object')
object_cols = list(obj[obj].index)
unique_values = {}
for col in object_cols :
   unique_values [col] = dataset[col]. unique().size
unique_values
```

```python
# Using Matplotlib and Seaborn Library for countplot the Category and Purpose Columns

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))

# Subplot 1
plt.subplot(1, 2, 1)
sns.countplot(data=dataset, x='CATEGORY')
plt.xticks(rotation=90)
plt.title('Countplot of CATEGORY')


```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sns.countplot(data = dataset, x = 'PURPOSE')
plt.xticks(rotation = 90)
plt.title('Countplot of Business')
```

```python
# for the timecolumn
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sns.countplot( data = dataset , x ='day-night')
plt.xticks(rotation = 90)
```

```python
# Comparing the two Different Categories along with 'PURPOSE' OF USER
plt.figure(figsize = (15,5))
sns.countplot( data = dataset, x = 'PURPOSE', hue = 'CATEGORY')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Insights from the above count - plots
# Most of the rides are booked for the business purpose
# Most of the people book cabs for Meetings and Meal / Entertain
# Most of the cabs are booked in the time duration ( 10 am - 5 pm )(Afternoon)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

object_cols = ['CATEGORY', 'PURPOSE']
OH_encoder = OneHotEncoder(sparse=False)
OH_cols = OH_encoder.fit_transform(dataset[object_cols])

# Get the column names for one-hot encoded columns
column_names = OH_encoder.get_feature_names_out(object_cols)

OH_cols = pd.DataFrame(OH_cols, columns=column_names, index=dataset.index)
df_final = dataset.drop(object_cols, axis=1)
dataset = pd.concat([df_final, OH_cols], axis=1)
dataset
```

```python
#Comment
plt.figure(figsize=(12, 6))
sns.heatmap(dataset.corr(),
            cmap='BrBG',
            fmt='.2f',
            linewidths=2,
            annot=True)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Insight from Heatmap
# 1. Business and Personal Category are highly negatively correlated.
#2. There is not much correlation between the features
```

```python
dataset['MONTH'] = pd.DatetimeIndex(dataset['START_DATE']).month
month_label = {1.0: 'Jan', 2.0: 'Feb', 3.0 : 'Mar', 4.0 : 'April',
               5.0: 'May', 6.0: 'June', 7.0: 'July', 8.0: 'Aug',
               9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0 : 'Dec'}

dataset["MONTH"] = dataset.MONTH.map(month_label)
mon = dataset.MONTH.value_counts(sort = False)
# Mon Total Rides count vs Month Ride Max Count
df = pd.DataFrame ({"MONTHS": mon.values,

                    "VALUE COUNT": dataset.groupby('MONTH', sort = False ) ['MILES'].max()})
p = sns.lineplot (data = df)
p.set(xlabel = "MONTHS", ylabel = "VALUE COUNT")

```
